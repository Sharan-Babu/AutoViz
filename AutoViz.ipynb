```python
import numpy as np
import pandas as pd
import autoviz
```

```python
# For this example, I am using an advertising data set. 
df = pd.read_csv(r"C:\Users\Sharan Babu\Desktop\advertising.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# Import the main package
from autoviz.AutoViz_Class import AutoViz_Class

# Instantiate a AutoViz object
analysis = AutoViz_Class()
```

```python
# Directly type your file location for the function to work. The function cannot use DataFrames.
report = analysis.AutoViz("C:/Users/Sharan Babu/Desktop/advertising.csv", depVar='Clicked on Ad',header=1, verbose=1)
```

Your visualization is done. You get scatter plots,Histograms and box plots among a few other visualizations.<br>
Done

Verbose =0 does not print any messages.
<br>
Verbose =1 print messages and displays charts on the terminal.
<br>
